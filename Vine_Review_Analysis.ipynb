```python
#import dependencies
import pandas as pd
import os

#retrieve file path to the vine_table.csv file
os.getcwd()
```

```python
#read the data into a dataframe
data_file_path = 'C:\\Users\\ssteffen\\desktop\\sam\\Vanderbilt Boot Camp\\MyRepo\\Module 16_Big_Data\\Amazon_Vine_Analysis\\vine_table.csv'
df = pd.read_csv(data_file_path)
df.head()
```

```python
# Filter the data and create a new DataFrame or table 
# to retrieve all the rows where the total_votes count is equal to or greater than 
# 20 to pick reviews that are more helpful and to avoid having division by zero errors later on.

filtered_df = df.loc[(df['total_votes'] >= 20)]
filtered_df
```

```python
# create a new DataFrame or table to retrieve all the rows where 
# the number of helpful_votes divided by total_votes is equal to or greater than 50%.
positive_df = filtered_df.loc[(filtered_df['helpful_votes']/filtered_df['total_votes'] >= .5)]
positive_df
```

```python
# Filter the DataFrame or table created in Step 2, 
# and create a new DataFrame or table that retrieves all the rows 
# where a review was written as part of the Vine program (paid), vine == 'Y'.
vine_df = positive_df.loc[(positive_df['vine'] == 'Y')]
vine_df.head()
```

```python
# Repeat Step 3, but this time retrieve all the rows where the review 
# was not part of the Vine program (unpaid), vine == 'N'.
no_vine_df = positive_df.loc[(positive_df['vine'] == 'N')]
no_vine_df.head()
```

```python
# Determine the total number of reviews, the number of 5-star reviews, 
# and the percentage of 5-star reviews for the two types of review (paid vs unpaid).
vine_reviews_total = vine_df['review_id'].count()
vine_five_star_reviews_total = vine_df.loc[(vine_df['star_rating'] == 5.0)]['review_id'].count()
vine_five_star_percentage = vine_five_star_reviews_total / vine_reviews_total * 100

no_vine_reviews_total = no_vine_df['review_id'].count()
no_vine_five_star_reviews_total = no_vine_df.loc[(no_vine_df['star_rating'] == 5.0)]['review_id'].count()
no_vine_five_star_percentage = no_vine_five_star_reviews_total / no_vine_reviews_total * 100

summary = (
    f"-------------------------\n"
    f"Vine Subscribers\n"
    f"-------------------------\n"
    f"Total Reviews: {vine_reviews_total}\n"
    f"Five Star Reviews: {vine_five_star_reviews_total}\n"
    f"Five Star Percentage: {vine_five_star_percentage:.2f}%\n"
    f"-------------------------\n"
    f"Non-Vine Subscribers\n"
    f"-------------------------\n"
    f"Total Reviews: {no_vine_reviews_total}\n"
    f"Five Star Reviews: {no_vine_five_star_reviews_total}\n"
    f"Five Star Percentage: {no_vine_five_star_percentage:.2f}%\n"
    f"-------------------------\n"
)

print(summary)
```

```python
# Determine the total number of reviews, the number of 5-star reviews, 
# and the percentage of 5-star reviews for the two types of review (paid vs unpaid).
vine_reviews_total = vine_df['review_id'].count()
vine_one_star_reviews_total = vine_df.loc[(vine_df['star_rating'] == 1.0)]['review_id'].count()
vine_one_star_percentage = vine_one_star_reviews_total / vine_reviews_total * 100

no_vine_reviews_total = no_vine_df['review_id'].count()
no_vine_one_star_reviews_total = no_vine_df.loc[(no_vine_df['star_rating'] == 1.0)]['review_id'].count()
no_vine_one_star_percentage = no_vine_one_star_reviews_total / no_vine_reviews_total * 100

summary = (
    f"-------------------------\n"
    f"Vine Subscribers\n"
    f"-------------------------\n"
    f"Total Reviews: {vine_reviews_total}\n"
    f"One Star Reviews: {vine_one_star_reviews_total}\n"
    f"One Star Percentage: {vine_one_star_percentage:.2f}%\n"
    f"-------------------------\n"
    f"Non-Vine Subscribers\n"
    f"-------------------------\n"
    f"Total Reviews: {no_vine_reviews_total}\n"
    f"One Star Reviews: {no_vine_one_star_reviews_total}\n"
    f"One Star Percentage: {no_vine_one_star_percentage:.2f}%\n"
    f"-------------------------\n"
)

print(summary)
```
